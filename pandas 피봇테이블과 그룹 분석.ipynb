2.성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다.

```python
titanic.groupby(["sex", "age_cat","class"])[['survived']].sum()/titanic.groupby(["sex","age_cat", "class"])[['survived']].count()
```

3.성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.

```python
titanic.pivot_table('survived', index=['sex', 'age_cat'], aggfunc='mean', columns='class')
```
